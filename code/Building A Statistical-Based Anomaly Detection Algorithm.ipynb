### Question 5) Model the same normal process “merged exp normal.csv” using Gaussian distribution with all requirements in Question 4. However, assume that the features are dependent [10 points].

Multivariate Gaussian Distribution of X1 and X5! First, we need to define our own functions to calculate the mean and variance of each feature, then calculate the joint Gaussian PDF.

##### Function that calculates and returns the mean and covariance of the data passed

```python
def estimateGaussian(X):
    n = np.size(X, 1)
    m = np.size(X, 0)
    mu = np.zeros((n, 1))
    sigma2 = np.zeros((n, 1))
    
    mu = np.reshape((1/m)*np.sum(X, 0), (1, n))
    sigma2 = np.reshape((1/m)*np.sum(np.power((X - mu),2), 0),(1, n))
    
    return mu, sigma2
```

##### Function that calculates and returns the Joint Multivariate Gaussian PDF for the data passed

```python
def multivariateGaussian(X, mu, sigma2):
     n = np.size(sigma2, 1)
     m = np.size(sigma2, 0)
     print("m = ", m, "\nn = ", n)
     
     if n == 1 or m == 1:        
         sigma2 = np.diag(sigma2[0, :])
         print("variance = ", sigma2)
     
     X = X - mu
     pi = math.pi
     det = np.linalg.det(sigma2)
     inv = np.linalg.inv(sigma2)
     val = np.reshape((-0.5)*np.sum(np.multiply((X@inv),X), 1),(np.size(X, 0), 1))
     
     p = np.power(2*pi, -n/2)*np.power(det, -0.5)*np.exp(val)
     
     print("X = ", X, " \nDET = ", det, " \nINV = ", inv, " \nVAL = ", val, " \nP = ", p)
     
     return p
```

##### Calculate mean and covariance of X1 and X5 as dependent features

```python
X = (((np.concatenate((x_one_normal.reshape(-1,1), x_five_normal.reshape(-1,1)), axis = 1))))
mu, sigma2 = estimateGaussian(X)
print('mean: ', mu,' variance: ', sigma2)
```

##### Buffer the output function (30 timesteps)

```python
def findAnomaliesWindow(binVec):
    i = 0
    while (i < len(binVec)-1):
        j = i
        while binVec[j] == 1:
            j+=1
        k = j - i
        if (k > 30):    
            print("ANOMALY WINDOW DETECTED. Indexes", i, "till", j)
            i+=k
        else:
            binVec[i:j] = 0
            i+=1
```

##### Function to return the best epsilon value to use for selecting outliers via predictions over a validation set. F1 score is used here to compare preditions produced with the test epsilon value instead of accuracy as the dataset is highly unbalanced.

```python
#Credits to Andrew Ng for this function. Code from https://www.coursera.org/learn/machine-learning
def selectThreshold(yval, pval):
    best_epi = 0
    best_F1 = 0
    
    stepsize = (max(pval) -min(pval))/1000
    epi_range = np.arange(pval.min(),pval.max(),stepsize)
    for epi in epi_range:
        predictions = (pval<epi)[:,np.newaxis]
        tp = np.sum(predictions[yval==1]==1)
        fp = np.sum(predictions[yval==0]==1)
        fn = np.sum(predictions[yval==1]==0)
        
        # compute precision, recall and F1
        prec = tp/(tp+fp)
        rec = tp/(tp+fn)
        
        F1 = (2*prec*rec)/(prec+rec)
        
        if F1 > best_F1:
            best_F1 =F1
            best_epi = epi
        
    return best_epi, best_F1
```

#### Network Architecture

There are various different deep learning network architectures out there, with each excelling in solving a particular type of problem. We have got a time-series forecasting problem in our hands, so a Recurrent Neural Network (RNN) is suitable to be used here as it models in the time dimension by creating recurrent cycles. The network includes a feedback loop for learning and an extra parameter for connections between time-steps, which allows them to excel in modelling temporal behaviour such as time-series, language, audio, and text. 

Given this, I have decided to implement an RNN with a many-to-one (8 to 1) Long Short Term Memory (LSTM) architecture. LSTM’s main components are the memory cells and the input, forget, and output gates. This allows them to have connections from previous time step as well as previous layers, where every output is influenced by the input as well as the historical inputs. There are multiple LSTM layers, and each layer is comprised of many LSTM units. The gates mentioned earlier are found in the LSTM units. The input gate protects the unit from irrelevant events, the forget gate helps the unit forget previous memory contents, and the output gates exposes the contents of the memory cell at the output of the LSTM unit. The following figure 1 shows the basic architecture of the LSTM network – 3 LSTM units in 1 LSTM layer.

##### Class to keep track of the model's history while training

```python
import time
class TimeHistory(Callback):
    def on_train_begin(self, logs={}):
        self.times = []

    def on_epoch_begin(self, epoch, logs={}):
        self.epoch_time_start = time.time()

    def on_epoch_end(self, epoch, logs={}):
        self.times.append(time.time() - self.epoch_time_start)
```

##### Function to calculate Root Mean Squared Error (RMSE)

```python
def rmse(y_true, y_pred):
    return backend.sqrt(backend.mean(backend.square(y_pred - y_true), axis=-1))
```

##### Define Model

The model was defined as sequential, as the LSTM layers were sequential to each other. LSTM layers were then added to the model. Each layer had the number of LSTM units, the recurrent activation function (Applied to gates to specify how much to activate them) and the activation function (Applied to update state and cell) specified, with the rest left as default. A Dropout layer with value of 0.2 was added after each LSTM layer, which is a regularization method where input and recurrent connections to LSTM units are probabilistically excluded from activation and weight updates while training the network. This helped in reducing overfitting and improving model performance. After all layers have been added, a Dense layer of value 1 was added, to make our model more robust and indicating a single value prediction in the output.

- The first parameter to the LSTM layer is the number of neurons or nodes that we want in the layer. 
- The parameter after the activations is the return_sequences, which is set to true since we will add more layers to the model. 
- The first parameter to the input_shape is the number of time steps while the last parameter is the number of indicators.

Hyperparameter values in this network were chosen by default due to their common and safe use in other worldwide LSTM models. 2 LSTM layers were selected as adding further layers did not really affect the average RMSE, MSE or MAPE, which makes sense as the power of LSTM comes mainly from its units rather than layers, and the more layers we got, the more time needed for model compilation and fitting.

Note: Because it is a 'binary classification' problem, log loss was used as the loss function (binary_crossentropy in Keras).

```python
model = Sequential()
model.add(LSTM(units=100, activation='softsign', recurrent_activation='sigmoid', return_sequences=True, input_shape=(train_X_lstm.shape[1], 8)))
model.add(Dropout(0.2)) #Regularization 
model.add(LSTM(units=100, activation='softsign', recurrent_activation='sigmoid'))
model.add(Dropout(0.2)) #Regularization
model.add(Dense(units = 1, activation='sigmoid'))
optimizerValue = Adam(learning_rate=0.001)
model.compile(optimizer = optimizerValue, loss = 'binary_crossentropy', metrics=['accuracy', 'mse', 'mape', rmse])
model.summary()
```

```python
#Memory Tracing
tracemalloc.start()

#LSTM Model Fit
time_callback = TimeHistory()
history = model.fit(train_X_lstm, train_Y_lstm, epochs = 100, batch_size = 16, callbacks = [time_callback])

#Print Memory Requirements
print(f"Current memory usage is {current / 10**6}MB; Peak was {peak / 10**6}MB")
tracemalloc.stop()
```
